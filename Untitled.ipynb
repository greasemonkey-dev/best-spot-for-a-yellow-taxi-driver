### Let's tag each row by its pickup time to identify the time of day ‚òÄÔ∏è

```python
def create_time_divisions(df):

  # Create a new column 'time_division' to categorize pickup times
  df['time_division'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.hour.apply(
      lambda x: 'Early Morning' if 0 <= x < 6 else (
          'Morning' if 6 <= x < 12 else (
              'Afternoon' if 12 <= x < 18 else 'Night'
          )
      )
  )
  return df
```

```python
df = create_time_divisions(df.copy())
```

### Tagging Weekday vs. Weekend Rides

```python
# Add weekday/weekend column based on pickup datetime
df_copy['week_day'] = pd.to_datetime(df_copy['tpep_pickup_datetime']).dt.weekday

# Define weekdays based on integer values (0 = Monday, 6 = Sunday)
weekdays = {0: 'Weekday', 1: 'Weekday', 2: 'Weekday', 3: 'Weekday', 4: 'Weekday', 5: 'Weekend', 6: 'Weekend'}

# Apply weekday labels based on the 'week_day' value
df_copy['week_day'] = df_copy['week_day'].map(weekdays)
```

```python
df_copy[['tpep_pickup_datetime','week_day']]
```

```python
df_copy['average_money_per_mile'] = df_copy.groupby(['PULocationID', 'time_division', 'week_day'])['money_per_mile'].transform(pd.Series.mean)
```

```python
ride_count = df_copy.groupby(['PULocationID', 'time_division', 'week_day'])['trip_distance'].size().reset_index(name='ride_count')
df_copy = pd.merge(df_copy, ride_count, on=['PULocationID', 'time_division', 'week_day'], how='left')
df_copy
```

```python
df_copy = df_copy.drop('ride_count_x', axis=1)  # Replace 'ride_count_x' with the actual column name
```

```python
df_copy['ride_count'] = df_copy['ride_count_y']  # Replace 'ride_count_y' with the actual column name
df_copy = df_copy.drop('ride_count_y', axis=1)  # Remove the renamed column (optional)
```

```python
average_fare_weight = 0.5
trip_count_weight = 0.5
df_copy['score'] = average_fare_weight * df_copy['average_money_per_mile'] + trip_count_weight * df_copy['ride_count']
```

# That's it üèÜ -  the *score* represents the best place for a taxi driver to wait (devided by part of the day in weekend/weekday

```python
grouped_df = df_copy.groupby(['PULocationID', 'time_division', 'week_day', 'score']).size().to_frame(name='count').sort_values(by='score', ascending=False)
grouped_df
```

```python
import matplotlib.pyplot as plt

# Create a bar chart
colors = ['pink' if day == 'weekend' else 'blue' for day in top_15_scores['week_day']]

ax = top_15_scores.plot(kind='bar', x='PULocationID', y='score', color=colors)
plt.xlabel('PULocationID')
plt.ylabel('score')
plt.title('Top 15 Highest Average Scores by PULocationID')
plt.legend(title='score')
plt.xticks(rotation=50)  # Rotate x-axis labels for better readability

# Add labels for weekday value and time division when hovering over the bar
for i, rect in enumerate(ax.patches):

    # Get the corresponding weekday value and time division from the DataFrame
    time_division = top_15_scores.iloc[i]['time_division']
    
    # Annotate the bar with the weekday value and time division
    ax.text(rect.get_x() + rect.get_width() / 2, rect.get_height(), f'{time_division}',
            ha='center', va='top', rotation=90,color='pink')

plt.show()
```
